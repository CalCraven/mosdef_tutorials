```python
%matplotlib notebook
from visualize import visualize
import mbuild as mb
```

```python
from mbuild.examples import Alkane

hexane = Alkane(6)
perfluorodecane = Alkane(10)
for particle in perfluorodecane.particles():
    if particle.name == 'H':
        particle.name = 'F'

visualize(hexane)
visualize(perfluorodecane)
```

```python
box = mb.Box(lengths=[5, 5, 5])

box_of_hexanes = mb.fill_box(compound=hexane, n_compounds=20, box=box)
visualize(box_of_hexanes)
```

```python
box_of_chains = mb.solvate(solute=box_of_hexanes, solvent=perfluorodecane, n_solvent=20, box=box)
visualize(box_of_chains)
```

```python
box_of_chains = mb.fill_box(compound=[hexane, perfluorodecane], n_compounds=[20, 20], box=box)
visualize(box_of_chains)
```

```python
box_of_chains = mb.fill_box(compound=[hexane, perfluorodecane], density=200, compound_ratio=[2, 1], box=box)
visualize(box_of_chains)
```
